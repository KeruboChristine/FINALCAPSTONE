## EDA
**Univariate Analysis**

```python
# IPC Phase distribution
plt.figure(figsize=(8,5))

df['description'].value_counts().plot(kind='bar')

plt.title('Distribution of IPC Food Security Phases')
plt.xlabel('IPC Phase')
plt.ylabel('Frequency')
plt.xticks(rotation=45)

plt.show()
```

Observation

The distribution is highly imbalanced.

Approximate frequencies are:

Minimal (Phase 1): ~14,300 observations
Stressed (Phase 2): ~9,100 observations
Crisis (Phase 3): ~4,100 observations
Emergency (Phase 4): ~200 observations
Famine (Phase 5): 0 observations
Interpretation

This indicates that Kenya spends most reporting periods in relatively stable food security conditions (Minimal and Stressed), while severe food insecurity events occur much less frequently.

This is expected because famine is an extremely rare event and FEWS NET assessments are designed to identify early deterioration before famine develops.

Research implication

This immediately tells us:

The dataset is class imbalanced.
Predicting Emergency conditions will be more difficult.
Evaluation should not rely only on accuracy if we frame the task as classification.

For classification, metrics such as Precision, Recall, and F1-score become more informative.




Shows :

Which IPC phase is the most common?

Is the dataset imbalanced?

```python
# TEMPORAL ANALYSIS
df['reporting_date'] = pd.to_datetime(df['reporting_date'])

df['Year'] = df['reporting_date'].dt.year
df['Quarter'] = df['reporting_date'].dt.quarter
df['Month'] = df['reporting_date'].dt.month
```

```python
#Observations pre year
year_counts = (
    df.groupby('Year')
      .size()
)

plt.figure(figsize=(12,5))

year_counts.plot(
    marker='o'
)

plt.title("Number of IPC Assessments per Year")

plt.xlabel("Year")
plt.ylabel("Number of Assessments")

plt.grid(True)

plt.show()
```

Observation

The number of assessments is relatively stable from 2012 to 2024.

There are slight decreases around 2016 and a noticeable reduction in 2025–2026.

Interpretation

The lower counts in 2025 and 2026 likely reflect incomplete data rather than fewer IPC assessments, as the dataset ends in early 2026.

Overall, the dataset has consistent temporal coverage over most of the study period.

Research implication

Consistent observations are essential for time-series forecasting because they reduce bias caused by irregular sampling.

```python
#average IPC Phase per year
year_mean = (
    df.groupby('Year')['value']
      .mean()
)

plt.figure(figsize=(12,5))

year_mean.plot(
    marker='o'
)

plt.title("Average IPC Phase by Year")

plt.xlabel("Year")
plt.ylabel("Average IPC Phase")

plt.grid(True)

plt.show()
```

Observation

Average IPC values fluctuate over time.

Notable increases occur around:

2017
2021–2023
2026

Lower averages are observed around 2014–2016.

Interpretation

These fluctuations suggest that food security conditions vary over time rather than remaining constant.

The increase in average IPC phases during some years may correspond to droughts, economic shocks, or other adverse events.

Research implication

This confirms that there is a temporal signal in the data, which is exactly what forecasting models aim to learn.

```python
# average IPC Phase by quarter
quarter_mean = (
    df.groupby('Quarter')['value']
      .mean()
)

plt.figure(figsize=(8,5))

quarter_mean.plot(
    kind='bar',
    edgecolor='black'
)

plt.title("Average IPC Phase by Quarter")

plt.xlabel("Quarter")
plt.ylabel("Average IPC Phase")

plt.show()
```

## Time Series Specific Analysis

```python
#Monthly Trend
monthly = (
    df.groupby('reporting_date')['value']
      .mean()
)

plt.figure(figsize=(14,6))

monthly.plot()

plt.title("Average IPC Phase Over Time")

plt.xlabel("Reporting Date")
plt.ylabel("Average IPC Phase")

plt.grid(True)

plt.show()
```

Observation

The average IPC phase varies considerably across the study period.

There are:

clear upward trends,
periods of decline,
several peaks and troughs.
Interpretation

The series is dynamic and non-stationary.

Research implication

For ARIMA, we will need to formally test for stationarity using tests such as the Augmented Dickey–Fuller (ADF) test and apply differencing if required.

For LSTM, the presence of nonlinear fluctuations is advantageous because LSTMs can learn such patterns.

```python
#rolling mean(4 Assessments)
rolling = monthly.rolling(4).mean()

plt.figure(figsize=(14,6))

plt.plot(monthly, label='Original')

plt.plot(
    rolling,
    linewidth=3,
    label='Rolling Mean (4)'
)

plt.title("Trend Using Rolling Mean")

plt.legend()

plt.show()
```

Observation

The rolling mean smooths short-term fluctuations while preserving the long-term trend.

The trend generally increases after around 2020.

Interpretation

The data contain both short-term variability and long-term movement.

Research implication

This confirms that the series contains meaningful temporal structure suitable for forecasting.
